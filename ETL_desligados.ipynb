verificando quais usuarios devem ser desativados no sistema, baseado nas planilhas enviadas pelo RH separadas por mes

```python
import pandas as pd
import numpy as np
```

```python
# Importação das tabelas dos funcionarios desligados
mes2 = pd.read_excel("/content/sample_data/02-2023.xlsx")
mes3 = pd.read_excel("/content/sample_data/03-2023.xlsx")

# Importação da relação de usuarios ativos retirados o sistema
Bsis = pd.read_excel("/content/sample_data/RSG005_UsuarioEmpresa.xlsx")
```

```python
# Unificando todas as panilhas dos funcioanrios desligados
colunas = ['Loja','Nome']
frame2 = mes2[colunas]
frame3 = mes3[colunas]
frame = pd.concat([frame2, frame3])

colunasSis = ['Loja', 'Nome', 'Identificador']
base = Bsis[colunasSis]
```

```python
#retirando espaços em branco
frame['Loja'] = frame['Loja'].str.strip()
frame['Nome'] = frame['Nome'].str.strip()
base['Loja'] = base['Loja'].str.strip()
base['Nome'] = base['Nome'].str.strip()
display(frame)
```

```python
#gerando quais usuarios tem em comum entre as duas fontes para ter a relação apenas dos que ainda precisa ser desativados
final = pd.merge(frame, base, how = 'inner', on = 'Nome')
```

```python
#baixando a planilha final com o resultado

final.to_excel("output.xlsx")
```
